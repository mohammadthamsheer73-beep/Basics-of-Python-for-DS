while loop

```python
#while loop reversing a number

num=123456
reverse=0

while num>0:
  digit=num % 10
  reverse=reverse*10+digit
  num=num//10

print(reverse)
```

```python
#armstrong number
num=91
count=0
result=0
temp=num

while temp> 0:
  count+=1
  temp//=10

temp=num
while temp >0:
  digit = temp % 10
  result+=digit**count
  temp=temp//10

if(result==num):
  print('the number is an Armstrong number')
else:
  print("not an Armstrong number")
```

for loop

```python
num=2
temp=str(num)
count=len(temp)
result=0

for digit in temp:
  result+= int(digit)**count

if(result==num):
  print('the number is an Armstrong number')
else:
  print("not an Armstrong number")
```
