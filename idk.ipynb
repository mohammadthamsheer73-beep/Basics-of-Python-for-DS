```python
class BankAccount:
    def __init__(self, owner, account_number, balance=0):
        self.owner = owner
        self.account_number = account_number
        self.balance = balance
        self.transactions = []  

    def deposit(self, amount):
        if amount <= 0:
            print(" Deposit amount must be positive.")
            return
        self.balance += amount
        self.transactions.append(f"Deposited: {amount}")
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print(" Amount must be positive.")
        elif amount > self.balance:
            print(" Insufficient balance.")
        else:
            self.balance -= amount
            self.transactions.append(f"Withdrawn: {amount}")
            print(f" {amount} withdrawn. New balance: {self.balance}")

    def transfer(self, amount, target_account):
        print(f" Transferring {amount} to {target_account.owner}...")
        self.withdraw(amount)
        if self.balance >= 0: 
            target_account.deposit(amount)
            self.transactions.append(f"Transferred {amount} to {target_account.owner}")

    def show_balance(self):
        print(f"\n Account: {self.owner} | No: {self.account_number}")
        print(f"    Balance: ₹{self.balance}")

    def show_history(self):
        print(f"\n Transaction History for {self.owner}:")
        if not self.transactions:
            print("   No transactions yet.")
        else:
            for i, t in enumerate(self.transactions, 1):
                print(f"   {i}. {t}")

def find_account(accounts, acc_number):
    for acc in accounts:
        if acc.account_number == acc_number:
            return acc
    return None

def main():
    accounts = []
    acc_counter = 1001  

    while True:
        print("\n=====  Bank Menu =====")
        print("1. Create Account")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Transfer")
        print("5. Check Balance")
        print("6. Transaction History")
        print("7. Exit")

        choice = input("Choose (1-7): ")

        if choice == "1":
            name = input("Enter account holder name: ")
            initial = float(input("Initial deposit (₹): "))
            acc = BankAccount(name, acc_counter, initial)
            accounts.append(acc)
            print(f" Account created! Account Number: {acc_counter}")
            acc_counter += 1

        elif choice in ("2", "3", "4", "5", "6"):
            acc_no = int(input("Enter your account number: "))
            acc = find_account(accounts, acc_no)

            if not acc:
                print(" Account not found.")
                continue

            if choice == "2":
                amt = float(input("Amount to deposit: "))
                acc.deposit(amt)

            elif choice == "3":
                amt = float(input("Amount to withdraw: "))
                acc.withdraw(amt)

            elif choice == "4":
                target_no = int(input("Enter target account number: "))
                target = find_account(accounts, target_no)
                if not target:
                    print(" Target account not found.")
                else:
                    amt = float(input("Amount to transfer: "))
                    acc.transfer(amt, target)

            elif choice == "5":
                acc.show_balance()

            elif choice == "6":
                acc.show_history()

        elif choice == "7":
            print(" Thank you for using the bank. Goodbye!")
            break

        else:
            print(" Invalid choice. Try again.")


main()







```
