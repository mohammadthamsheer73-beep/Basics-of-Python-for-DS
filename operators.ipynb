#comparision operator

```python
a=5
b=6
if(a==b):
  print("a and b are equal")
if(a<b):
  print("a is lesser than b")
if(a>b):
  print("a is greater than b")
if(a<=b):
  print("a is lesser than equal to b")
if(a>=b):
  print("a is greater than equal to b")
if(a!=b):
  print("a and b are not equal")
```

logical operator

```python
x=25
if(x>10 and x>20):
  print('x is a large number')
if(x>10 or x>30):
  print("x is either larger than 10 or 30")
if not(x>256):
  print("x is really small")
```
