```python
#number guessing game without random
x=12
attempts=0
while(True):
    y=int(input("enter a number between 1 to 100:"))
    if y<1 or y>100:
        print("invalid")    
        continue
    attempts +=1
    if (y==x):
        print("congrats!")
        print(f"you tried {attempts} times")
        break
    elif(y>x):
        print("too high")
    else:
        print("too low")
```
