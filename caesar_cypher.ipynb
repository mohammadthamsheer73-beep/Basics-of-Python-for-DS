```python
String=input("enter a string:")
temp=0
b=[]
for x in String:
    c=ord(x)

    if c>=ord("A") and c <= ord("Z"):
        c=(c-ord("A")+3) % 26 + ord("A")
        temp=chr(c)
        b.append(temp)
    elif c>=ord("a") and c<=(ord("z")):
        c=(c-ord("a")+3)%26+ord("a")
        temp=chr(c)
        b.append(temp)
    else:
        b.append(x)
print(b)
```
