```python
String=input("enter a string:")
temp=0
b=[]
for x in String:
    c=ord(x)

    if c>ord("W") and c < ord("a"):
        c=(c-ord("A")+3)%26+ord("A")
        temp=chr(c)
        b.append(temp)
    if c>=ord("w"):
        c=(c-ord("a")+3)%26+ord("a")
        temp=chr(c)
        b.append(temp)
    else:
        temp=chr(c+3)
        b.append(temp)
print(b)
```
